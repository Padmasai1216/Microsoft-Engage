```python
#remove columns that are 90% emmpty
threshold=len(google_data)*0.1
threshold
```

```python
google_data.dropna(thresh = threshold,axis=1,inplace=True)
```

```python
#Data imputation and manipulation
#filling null values with mean , median and mode\
def impute_median(series):
    return series.fillna(series.median())
```

```python
google_data.Cylinders=google_data['Cylinders'].transform(impute_median)
```

```python
print(google_data.isnull().sum())
```

```python
#filling categorical values with mode

print(google_data['Make'].mode())
```

```python
print(google_data['Paddle_Shifters'].mode())
```

```python
print(google_data['Cruise_Control'].mode())
```

```python
google_data['Make'].fillna(str(google_data['Make'].mode().values[0]),inplace=True)
google_data['Paddle_Shifters'].fillna(str(google_data['Paddle_Shifters'].mode().values[0]),inplace=True)
google_data['Cruise_Control'].fillna(str(google_data['Cruise_Control'].mode().values[0]),inplace=True)
```

```python
#data visualization

grp=google_data.groupby('Make')
x=grp['Cylinders'].agg(np.mean)
y=grp['Seating_Capacity'].agg(np.sum)
z=grp['Doors'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x,'ro')
plt.xticks(rotation=90)
plt.title('Make wise Cylinders')
plt.xlabel('Make-->')
plt.ylabel('Cylinders-->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(y,'r^',color='b')
plt.xticks(rotation=90)
plt.title('Make wise Seating_Capacity')
plt.xlabel('Make-->')
plt.ylabel('Seating_Capacity-->')
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.plot(z,'g--', color='g')
plt.xticks(rotation=90)
plt.title('Make wise Doors')
plt.xlabel('Make-->')
plt.ylabel('Doors-->')
plt.show()
```
